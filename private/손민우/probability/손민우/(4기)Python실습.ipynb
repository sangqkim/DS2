## 분포 소개
 1) 확률분포(probability distribution) : 확률변수 X가 가질 수 있는 값과 해당하는 확률에 대해 나타낸 것
<br>
$$Pr(a<x<b)=\int_{a}^{b}f(x)dx$$

2) 기대값(expected value) : 


- $\mu=E(X)=\int_{-\infty}^{\infty}xf(x)dx$

3) 분산(varianve)과 표준편차(standard deviation) :


- $Var(X)=E[(X-\mu)^2]=E(X^2)-[E(X)]^2$


- $sd(X)=\sqrt{Var(X)}$

4) 이항분포 :$X\sim B(n,p)$
 - $X_1,X_2,\dots,X_n$은 서로 독립이고, 각각 성공률이 p인 베르누이 확률변수
 
 
 - $X=X_1+X_2+\dots,+X_n$은 성공률이 p인 베르누이 시행을 n번 반복시행한 결과로 X는 n번 시행할 때의 성공횟수이다.
 
 
 - $f(x)=\binom{n}{x}p^x(1-p)^{n-x}$


5) 균일분포 : $X \sim U(a,b)$
 - 확률변수 X가 a와 b사이에서 동일한 값을 가질 때 X를 균일분포를 따른다고 한다.
 
 
 - $f(x)=\frac{1}{b-a},a<x<b$

6) 정규분포 : $X \sim N(\mu,\sigma^2)$
 - 평균과 표준편차에 의해 결정되는 분포로 종형의 확률밀도함수를 갖는다.
 
 
 - $f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}exp(-\frac{(x-\mu)^2}{2\sigma^2})$
 
7) 카이제곱분포 : $\sum_{i=1}^{k}Z_i^2\sim \chi^2(k)$
- 확률변수 $Z_1,Z_2,\dots,Z_k$ 이 정규분포 $N(0,1)$로부터 추출한 k개의 랜덤표본일 때, $V=Z_1^2+Z_2^2\dots+Z_k^2$ 의 분포를 자유도 k인 카이제곱분포라고 한다.


- 정규분포 $N(\mu,\sigma^2)$로부터 추출한 n개의 랜덤표본 $X_1,X_2,\dots,X_n$ 의 표본분산 $S^2=\frac{1}{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2$ 에 대해 다음이 성립한다.
<br>

$$\frac{(n-1)S^2}{\sigma^2} \sim \chi^2(n-1)$$

8) t분포 : $T \sim t(k)$
- 표준정규분포를 따르는 확률변수를 Z라 하고, 이와는 독립이며 자유도가 k인 카이제곱분포를 따르는 확률변수를 V라고 할 때, $T=\frac{Z}{\sqrt{V/k}}$ 의 분포를 자유도가 k인 t분포라고 한다.


- $X_1,X_2,\dots,X_k$ 이 정규분포 $N(\mu,\sigma^2)$를 따르는 정규모집단에서 추출한 랜덤표본일 때, 스튜던트화된 표본평균의 분포가 다음과 같다.
<br>

$$\frac{\bar{X}-\mu}{S/\sqrt{n}} \sim t(n-1)$$

9) F분포 : $F \sim F(k_1,k_2)$
- F분포는 두 정규모집단의 분산을 비교할 때 분산의 비에 대한 추론에 이용되는 분포이다.


- 두 확률변수 $V_1,V_2$ 가 서로 독립이고 각각 자유도가 $k_1,k_2$인 카이제곱 분포를 따를 때, $F=\frac{V_1/k_1}{V_2/k_2}$ 의 분포를 자유도가 $(k_1,k_2)$인 F분포라고 한다.


- 서로 독립인 두 정규모집단 $N(\mu_1,\sigma_1^2)$, $N(\mu_2,\sigma_2^2)$에서 추출한 랜덤표본 $X_1,X_2,\dots,X_{n_1}$, $Y_1,Y_2,\dots,Y_{n_2}$, 그리고 각각의 표본분산 $S_1^2,S_2^2$ 에 대해 다음과 같은 식이 성립한다.
<br>

$$\frac{\sigma_2^2}{\sigma_1^2}.\frac{S_1^2}{S_2^2} \sim F(n_1-1,n_2-1)$$


- 확률변수 F가 자유도가 $(k_1,k_2)$인 F분포를 따를 때, $\frac{1}{F} \sim F(k_2,k_1)$ 가 성립한다.

scipy.stats.binom.pmf(x,n,p) : scipy 패키지만 불러온 경우
 
binom.pmf(x,n,p) : scipy.stats 에서 binom 모듈을 따로 불러온 경우
 
### 이항분포의 확률밀도 - binom.pmf(x,n,p)
- x:이항분포를 따르는 확률변수값
  
  n:이항분포에서의 총 베르누이 시행횟수
  
  p:베르누이 시행 시 성공확률


- $X\sim B(n,p)$ 일 때, $Pr(X=x)={{n}\choose{x}}P^x (1-P)^{n-x}$의 값을 구한다.

### 이항분포의 누적확률 - binom.cdf(k,n,p)
- k:0부터 k까지의 누적확률을 구한다.


- $X \sim B(n,p)$일 때, $Pr(X\leq k)=\sum_{i=0}^{k}{{n}\choose{i}}P^i(1-P)^{n-i}$의 값을 구한다.

### 이항분포의 백분위수 - binom.ppf(q,n,p)
- q:구하고 싶은 분위수


- $X \sim B(n,p)$일 때, $Pr(X\leq x)=q$인 x의 값을 구한다.

### 이항분포의 난수생성 - binom.rvs(n,p,size)
- size:구하고 싶은 난수의 개수


- $X \sim B(n,p)$에서 'size'개의 난수를 생성한다.

```python
### 어느 제품의 불량률은 10% 이다. 임의의 5개의 제품 중에 불량품이 3개일 확률은?
import scipy
from scipy.stats import binom

print(round(binom.pmf(3,5,0.1),4))
# round(수,자릿수):반올림함수

### 어느 제품의 불량율은 10%이다. 임의의 20개의 제품 중에 불량품이 5개 이하일 확률은?
print(round(binom.cdf(k=5,n=20,p=0.1),4))
# print 함수는 강의자료에서 여러 결과를 나타내기 위해 사용한 것으로 생략 가능하다.

## 불랑품이 4개 이상일 확률은?
print(round(1-binom.cdf(k=3,n=20,p=0.1),4))
```

```python
print(binom.ppf(q=0.1875,n=5,p=0.5))

print(binom.rvs(5,0.5,size=3))
# size 옵션은 꼭 명시해주어야 한다.
```

이항분포와 마찬가지로 from scipy.stats import uniform 을 통해 균일분포 함수들을 사용할 수 있다. python에서는 location 변수와 scale 변수를 통해 Uniform(loc,loc+scale)의 형태로 균일분포를 표현한다. $X \sim U(a,b)$ 와 같이 표현한다.


### 균일분포의 확률밀도 - uniform.pdf(x,loc,scale)
- x:균일분포를 따르는 확률변수값


- loc:균일분포에서 하한값(a)


- sclae:균일분포에서 상한값-하한값(b-a)

### 균일분포의 누적확률 - uniform.cdf(x,loc,scale)
- x:여기서의 x 값은 확률변수값 x까지의 누적확률을 구한다는 의미이다.

### 균일분포의 백분위수 - uniform.ppf(q,loc,scale)
- q:구하고 싶은 분위수

### 균일분포의 난수생성 - uniform.rvs(loc,scale,size)
- size: $X \sim U(a,b)$에서 size개의 난수를 생성한다.

```python
from scipy.stats import uniform

print(uniform.pdf(x=1,loc=0,scale=2))

print(uniform.cdf(x=1,loc=0,scale=2))

print(uniform.ppf(q=0.3,loc=0,scale=2))

print(uniform.rvs(loc=0,scale=2,size=2))
```

from scipy.stats import chi2 코드를 이용하여 카이제곱분포 함수들을 사용할 수 있다.
### 카이제곱분포의 확률밀도 - chi2.pdf(x,df)
- x : 카이제곱분포를 따르는 확률변수값


- df : 자유도

### 카이제곱분포의 누적확률 - chi2.cdf(x,df)
- $X \sim \chi^{2}(df)$의 확률밀도함수 $P(X \leq x)$의 값을 구한다.

### 카이제곱분포의 백분위수 - chi2.ppf(q,df)
- $X \sim \chi^{2}(df)$일 때, $P(X \leq x)=q$를 만족하는 x의 값을 구한다.

### 카이제곱분포의 난수생성 - chi2.rvs(df,size)
- size개 만큼의 난수를 생성한다.

```python
from scipy.stats import chi2

x=np.arange(0,40.1,0.1)

plt.plot(x,chi2.pdf(x,3), color='black', label='df 3')
plt.plot(x,chi2.pdf(x,6),color='rosybrown', label='df 6')
plt.plot(x,chi2.pdf(x,9),color='seagreen', label='df 9')
plt.plot(x,chi2.pdf(x,20),color='blueviolet', label='df 20')

plt.legend(loc='upper right') 
#우측 상단에 범례
# plt.plot의 label 옵션을 통해 범례에 사용할 선의 이름을 지정할 수 있다.
```

### 모평균에 대한 유의성 검정(모분산을 모르는 정규모집단) 


전구를 생산하는 한 회사에서 현재 생산하는 전구의 평균수명은 1950시간으로 알려져 있다. 개발 중인 전구의 평균수명이 기존의 전구보다 수명이 더 길다고 할 수 있는지 판단하기 위해 9개의 시제품을 생산하여 그 수명시간을 조사한 결과가 다음과 같다. {2000,1975,1900,2000,1950,1850,1950,2100,1975} 수명의 분포가 정규분포라는 전제하에서 가설에 대한 유의수준 5%의 검정을 해보자.

- 가설 $H_0 : \mu=1950$, $H_1 : \mu>1950$



- 유의수준 $\alpha=0.05, n=9 $


- 검정통계량 $T=\frac{\bar{X}-\mu_0}{S/\sqrt{n}}\sim t(8)$ under $H_0$



- spicy 패키지의 stats 서브패키지에서 다양한 통계적 검정함수를 제공한다.


- import scipy.stats as st 명령어를 통해 서브패키지를 저장한 후 간편하게 사용한다.


- 단, python에서 지원하는 검정함수에서는 단측가설을 지원하지 않기 때문에, 양측가설의 결과를 통해 단측가설의 결과를 추정해야 한다.


- 일반적으로 대칭 분포를 따르는 모집단에 대한 검정에서 단측가설의 유의확률은 양측가설의 유의확률의 1/2로 나타난다.

```python
bulb=np.array([2000, 1975, 1900, 2000, 1950, 1850, 1950, 2100, 1975])

import scipy.stats as st

tt=st.ttest_1samp(bulb,1950) 
# ttest_1samp : 일표본 t검정 함수
print(round(tt.statistic,3)) 
# ttest_1samp를 이용한 검정통계량

m=np.mean(bulb)
#표본평균 

s=np.std(bulb)*np.sqrt(9/8)
#표본표준편차 - numpy에서 지원하는 분산 및 표준편차는 모분산과 모표준편차의 값을 제공한다. 표본분산과 표본표준편차의 값을 얻기 위해서는 각각 n/n-1, sqrt(n/n-1)을 곱해주어야 한다.

print(round((m-1950)/(s/np.sqrt(9)),3))
# 식을 이용해 직접 계산한 검정통계량. 둘의 값은 같다.

print(round(tt.pvalue/2,3)) 
# ttest_1samp를 이용한 유의확률

print(round(1-t.cdf(tt.statistic,8),3))
# 식을 이용해 직접 계산한 유의확률. 둘의 값은 같다.

# 유의확률이 0.05보다 크므로 귀무가설을 기각할 수 없다.
```

### 독립 이표본에 의한 모평균의 비교

$11-OH-\tiny{\triangle}^9 THC$ : {15.95,25.89,20.53,15.52,14.18,16.00}
  
|    <br>  $\text{}$   | <br>$\tiny{\triangle}^9 THC$    |  <br>  $11-OH-\tiny{\triangle}^9 THC$ &nbsp;&nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  |

표본표준편차의 차이가 크지 않다고 판단, 등분산을 가정한다.

- 가설 $H_0 : \mu_1=\mu_2$, $H_1 : \mu_1 \neq \mu_2$


- 합동표본분산 ${S_p}^2=\frac{(n_1-1){S_1}^2+(n_2-1){S_2}^2}{n_1+n_2-2}$


- 검정통계량 $T=\frac{(\bar{x_1}-\bar{x_2})-0}{S_p \sqrt{\frac{1}{n_1}+\frac{1}{n_2}}}\sim t(n_1+n_2-2)$ under $H_0$

- 가설 $H_0 : \mu_A=\mu_B$, $H_1 : \mu_A \neq \mu_B$


- 검정통계량 $T=\frac{(\bar{x_A}-\bar{x_B})-0}{\sqrt{\frac{{S_A}^2}{n_A}+\frac{{S_B}^2}{n_B}}}\sim t(df^*)$ under $H_0$

# 7장. 베이지안 분석
## 1. Metropolis-Hastings

### 목적 
- 원하는 분포$\left( P(x) \right)$에서 direct sampling이 어려울 때, 그 분포를 따르는 random samples 얻기 위해서
- Approximate the desired distribution (histogram)
- Compute an integral (ex : 기댓값 계산)

### Algorithm
$1.$ 초기값 설정 $x_0$, $\;t=0$

$2.$ 제안분포 $g(.)$에서 다음 단계의 후보값 $x^{\prime}$을 랜덤하게 뽑는다.

$3.$ Acceptance probability$\left(A\left(x^{\prime}, x_{t}\right)=\min \left(1, \frac{P\left(x^{\prime}\right)}{P\left(x_{t}\right)} \frac{g\left(x_{t} | x^{\prime}\right)}{g\left(x^{\prime} | x_{t}\right)}\right) \right)$를 계산한다.

$4.$ Accept of Reject 결정 : 
$$
\begin{array}{l}{\text { [1] generate a uniform random number } u \in[0,1] \text { ; }} \\ {\text { [2] if } u \leq A\left(x^{\prime}, x_{t}\right), \text { accept the new state and set } x_{t+1}=x^{\prime} \text { ' }} \\ {\text { [3] if } u>A\left(x^{\prime}, x_{t}\right), \text { reject the new state, and copy the old state forward } x_{t+1}=x_{t} \text { ; }}\end{array}
$$

$5.$ Set $t = t+1$, $\;\; 2\sim 5$ iterate

- 다음 상태로 나아갈지에 대한 여부를 acceptance probability에 근거하여 결정하는 알고리즘  (Sometimes accepting the moves and sometimes remaining in place)


- Acceptance ratio $\alpha$의 의미 : 새로운 후보값이 현재값에 비해 $P(x)$에서 나온 sample일 가능성 (>1 : 후보값이 현재값보다 원하는 분포에서 나왔을 가능성이 더 높다.) 


- 그러므로, 단계가 반복될수록 $P(x)$의 high-density regions에서 sample을 뽑게 된다. 

### 예시 1
#### 어느 공정에 하루동안 투입되는 화학물질의 양의 100일간 데이터가 있다. 이 데이터는 정규분포를 따르고, 과거 정보에 의하면 표준편차가 1이라고 알려져있다. 이 때 화학물질이 투입되는 일일 평균양에 대한 사후분포를 M-H 방법으로 구해보자.



- $Y_1, ...,Y_{100} \sim i.i.d \; N(\theta,1)$의 데이터가 존재  ($\theta = 3$)


- $\theta$의 사전분포 : $\pi_0(\theta)=\frac{1}{\pi(1+\theta^2)}$


- 사후분포 : 
$$
\begin{aligned} \pi\left(\theta | Y_{1}, \cdots, Y_{n}\right) \; & \propto \exp \left(-\frac{\sum_{i=1}^{n}\left(Y_{i}-\theta\right)^{2}}{2}\right) \times \frac{1}{1+\theta^{2}} \\ & \propto \exp \left(-\frac{n(\theta-\overline{Y})^{2}}{2}\right) \times \frac{1}{1+\theta^{2}} \end{aligned}
$$



- 제안분포 : $q\left(\theta | \theta^{*}\right)=\frac{1}{\sqrt{2 \pi n^{-1}}} e^{-\frac{(\theta-\overline{Y})^{2}}{2 / n}}, \text { i.e. } N\left(\overline{Y}, \frac{1}{n}\right)$


- acceptance ratio $\alpha$ : 
$$
\begin{aligned} \alpha\left(\theta^{(i-1)}, \tilde{\theta}\right) &=\min \left(1, \frac{\pi\left(\tilde{\theta} | Y_{1}, \cdots, Y_{n}\right) q\left(\theta^{(i-1)} | \tilde{\theta}\right)}{\pi\left(\theta^{(i-1)} | Y_{1}, \cdots, Y_{n}\right) q\left(\tilde{\theta} | \theta^{(i-1)}\right)}\right) \\ &=\min \left(1, \frac{1+\left(\theta^{(i-1)}\right)^{2}}{1+(\tilde{\theta})^{2}}\right) \end{aligned}
$$

```python
import scipy
from scipy.stats import norm
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)
# Observation of 100 individual, with mean = 3 and scale = 1 (실습에서는 데이터를 generate해서 사용하자.)
n = 100
observation = norm.rvs(3, 1, n)
```

```python
# 관측한 data의 histogram
plt.figure(figsize = (10, 10))
plt.hist(observation, bins = 25, )
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of 100 obs sampled from a pop of 10,000 with mu=3, sigma=1")
```

```python
mu_obs = observation.mean()
mu_obs
```

```python
# setting
para_init = 1
iterations = 10000
### Metropolis-Hastings algorithm ###
theta = []
theta_old = para_init
for i in range(iterations):
    theta_new = norm.rvs(mu_obs, 1/n, 1)
    log_alpha = min(np.log(1), np.log(1+theta_old**2)-np.log(1+theta_new**2))
    r = uniform.rvs(0, 1, 1)
    if np.log(r) < log_alpha:
        theta_old = theta_new
        theta.append(theta_new)
    else :
        theta.append(theta_old)
```

```python
# output
len(theta)
```

```python
# 앞의 9000개는 burn-in period, 나머지로 정상분포 추정
theta_burn = theta[9000:] 
# histogram of MCMC samples
plt.figure(figsize = (10, 10))
plt.hist(np.array(theta_burn), bins = 30, )
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of theta")
```

## 이항분포와 베타사전분포
- $y \sim Bernoulli(\theta) = Binomial(1,\theta)$

- The pdf is $$p(y|\theta)=\theta^y(1-\theta)^{1-y} = \left\{\begin{array}{ll}{\theta} & {y=1} \\ {1-\theta} & {y=0}\end{array}\right.$$

### Likelihood function
$$p(Y|\theta) = \prod_{i=1}^{T}\theta^{y_i}(1-\theta)^{1-y_i} = \theta^{s}(1-\theta)^{T-s}$$ where $s = \sum_i y_i$

### Prior
$\theta$는 확률 값이기 때문에 $\theta \in (0,1)$에 맞는 사전분포를 고려해야 한다.

- (conjugate) Beta prior $\theta \sim Beta(\alpha_1, \alpha_2)$ 

$$p(\theta;\alpha_1,\alpha_2)=\frac{1}{B(\alpha_1,\alpha_2)}\theta^{\alpha_1-1}(1-\theta)^{\alpha_2-1}$$

### Posterior
앞에서 가정한 likelihood function과 prior 분포를 통해 posterior분포를 유도할 수 있다.

<br>

$$
\begin{aligned} p\left(\theta | Y ; \alpha_{1}, \alpha_{2}\right) &=\frac{P(Y | \theta) P(\theta)}{P(Y)} \\ & \propto P(Y | \theta) P(\theta) \\ &=\theta^{s}(1-\theta)^{T-s} \frac{1}{B\left(\alpha_{1}, \alpha_{2}\right)} \theta^{\alpha_{1}-1}(1-\theta)^{\alpha_{2}-1} \\ & \propto \theta^{s+\alpha_{1}-1}(1-\theta)^{T-s+\alpha_{2}-1} \end{aligned}
$$

- Beta posterior $\theta|y \sim Beta(s+\alpha_1, T-s+\alpha_2)$

### 예시 2 
#### 어느 축구선수의 지난 100번의 경기에 대해 각 경기에서 골을 넣었는지 여부를 기록한 자료가 있다. 자료를 통해 이 선수가 경기에서 골을 넣을 확률에 대하여 베이지안 추정을 해보자.


- $Y_1, ...,Y_{100} \sim i.i.d \; Bernoulli(\theta) \; (0\le \theta \le 1)$의 데이터가 존재 ($\theta = 0.7$)

- 제안분포 :  $q\left(\theta | \theta^{*}\right)=\frac{1}{\sqrt{2 \pi (0.03)^2}} e^{-\frac{(\theta-\theta^{*})^{2}}{2 (0.03)^2}}, \text { i.e. } N\left(\theta^{*}, (0.03)^2\right)$

- symmetric proposal distribution $q\left(\theta | \theta^{*}\right)=q\left(\theta^{*} | \theta\right)$


- Acceptance ratio : 
$$
\begin{aligned}
\alpha(\theta, \theta^{*}) & =\min \left(1, \frac{p(\theta^{*}|Y) q(\theta | \theta^{*})}{p(\theta |Y) q(\theta^{*} | \theta)}\right) \\ & = \min \left(1, \frac{p(Y | \theta^{*}) p(\theta^{*}) q(\theta | \theta^{*})}{p(Y|\theta)p(\theta) q(\theta^{*} | \theta)}\right) \\ & = \min \left(1, \frac{p(Y | \theta^{*}) p(\theta^{*})}{p(Y|\theta)p(\theta)}\right) 
\end{aligned}
$$

```python
# Simulate data
np.random.seed(2019)

nobs = 100
theta = 0.7
Y = np.random.binomial(1, theta, nobs) # Observed sample form bernoulli(0.7)

# For alpha_1 = alpha_2 = 1, the Beta distribution (= uniform)
a1 = 1
a2 = 1
# parameters of posterior dist
a1_hat = a1 + Y.sum()
a2_hat = a2 + nobs - Y.sum()
```

```python
iterations = 10000 # draws from the (converged) posterior

# Model parameters
sigma = 0.03
para_init = 0.5             # initial value for theta

# Callable functions for likelihood and prior
mh_ll = lambda theta: Y.sum()*np.log(theta) + (nobs-Y.sum())*np.log(1-theta)
def mh_lprior(theta):
    lprior = (a1-1)*np.log(theta)+(a2-1)*np.log(1-theta)
    return lprior
mh_log_foralpha = lambda theta: mh_ll(theta) + mh_lprior(theta)
```

```python
### Metropolis-Hastings algorithm ###
thetas = []
theta = para_init
for i in range(iterations):
    # Draw theta
    # Generate the proposal
    etas = np.random.normal(0, sigma, 1)  # random walk errors
    theta_star = theta + etas # 체인의 다음상태에 대한 후보 theta_star 생성 (symmetric proposal distribution)
   
    # acceptance probability
    theta_prob = mh_log_foralpha(theta)
    theta_star_prob = mh_log_foralpha(theta_star) 
    log_accept_prob = min(0, theta_star_prob - theta_prob)
    
    # Append the new draw
    unif = np.random.uniform(size = 1)
    if log_accept_prob > np.log(unif):
        theta = theta_star
        theta_prob = theta_star_prob
        thetas.append(theta)
    else :
        thetas.append(theta)
```

```python
# 사후기댓값의 추정량
print(round(np.array(thetas[9000:]).mean(), 4)) # 한 경기에서 골을 넣을 확률은 약 66%
# 사후중앙값의 추정량
print(round(np.median(np.array(thetas[9000:])), 4))
# credible interval
print(round(np.percentile(thetas[9000:], 2.5), 4)) # 2.5% 분위수
print(round(np.percentile(thetas[9000:], 97.5), 4)) #97.5% 분위수
```

```python
# Plot the posterior
plt.figure(figsize=(10,4))
# Plot MH draws
plt.hist(np.array(thetas[1000:]), bins=50, density=True);
# Plot analytic posterior
X = np.linspace(0,1, 1000)
plt.plot(X, stats.beta(a1_hat, a2_hat).pdf(X), 'r');
# Plot prior (uniform dist)
plt.plot(X, stats.beta(a1, a2).pdf(X), 'g')

plt.title('Metropolis-Hastings via Python (10,000 Draws)')
plt.ylim(0, 12)
plt.legend(['Posterior (Analytic)', 'Prior', 'Posterior Draws (MH)']);
```

## 2. 깁스 샘플링(Gibbs Sampling)
### 특징
- 2차원 이상의 모수에 대한 사후분포를 각 모수에 대한 조건부 사후분포의 집합으로 나타낸다.

$\quad$ ex) $\,$2차원 모수 $\Theta=(\mu, \sigma)$에 대한 사후분포 $\pi(\mu, \sigma | Y_{1}, \cdots, Y_{n})$를 $\{\pi(\mu| \sigma, Y_{1}, \cdots, Y_{n}),\; \pi(\sigma | \mu, Y_{1}, \cdots, Y_{n})\}$으로 나타낸다. 

- joint distribution을 구하기 어려울 때나 이 분포에서 바로 sampling하기 어려울 때 각 모수에 대한 conditional distribution을 안다면 적용 가능

- algorithm은 다른 변수의 현재값에 따라 각 변수의 분포에서 차례로 샘플을 생성한다. (즉, 생성된 samplie sequence는 Markov Chain을 구성)

- Markov Chain의 정상분포는 joint distribution (사후분포)

$$ $$

*참고) Conditional distribution과 Joint distribution의 관계 

: 한 변수에 대한 conditional distribution(다른 변수들은 고정된 상태)은 joint distribution에 비례한다.

<br>

$$p\left(x_{j} | x_{1}, \ldots, x_{j-1}, x_{j+1}, \ldots, x_{n}\right)=\frac{p\left(x_{1}, \ldots, x_{n}\right)}{p\left(x_{1}, \ldots, x_{j-1}, x_{j+1}, \ldots, x_{n}\right)} \propto p\left(x_{1}, \ldots, x_{n}\right)$$

### 예시
#### 어떤 공장의 각 Lot별 생산된 wafer들의 수율은 정규분포를 따른다고 한다. 50개의 Lot을 랜덤 추출하여 Lot의 수율에 대한 평균과 분산을 Gibbs Sampling 방법으로 베이지안 추정을 해보자.

- $X_{1}, \cdots, X_{n=50} \sim \text { i.i.d } N\left(\mu, \sigma^{2}\right) $의 데이터가 존재 ($\mu = 95(\%), \sigma^2 = 1(\%)$)

- 목적 : $\mu, \sigma^2$ 추정하는 것

- 사전분포 : $\pi(\mu) \propto 1 \;\; \& \;\; \pi(\sigma^2) \propto \frac{1}{\sigma^2}$

- 조건부 사후분포 :
 $$\begin{aligned} \mu | \boldsymbol{X}, \sigma^{2} & \sim N\left(\overline{X}, \frac{\sigma^{2}}{n}\right) \\ \sigma^{2} | \boldsymbol{X}, \mu & \sim \text { IGamma }\left(\frac{n}{2}, \frac{1}{2} \sum_{i=1}^{n}\left(X_{i}-\mu\right)^{2}\right) \end{aligned}$$
 
- algorithm 

$
\begin{array}{l}{\quad\text { [1] 초기값 } \left( \mu^{(0)}, \sigma^{2(0)}\right) \text { 을 정한다. }} \\ {\quad\quad \text{s = 1 부터 N까지 다음을 반복}} \\ {\quad \text { [2] } \mu^{(s)} \sim N\left(\overline{X}, \frac{\sigma^{2(s-1)}}{n}\right)}\\ {\quad \text{ [3] }  \sigma^{2(s)} \sim \text { IGamma }\left(\frac{n}{2}, \frac{1}{2} \sum_{i=1}^{n} \left(X_{i}-\mu^{(s)}\right)^{2}\right)} \end{array}
$

초기값으로 $\mu^{(0)} = 95\%, \sigma^{2(0)} = 0.3 \%$로 놓고, $N = 10,000$개의 Gibbs sampler $\left(\left(\mu^{(s)}, \sigma^{2(s)}\right), s=1, \cdots, 10000\right)$를 생성

```python
# 평균 95, 표준편차 1인 정규분포를 따르는 50개의 데이터
n = 50
obs = norm.rvs(95, 1, n)

# 관측한 data의 histogram
plt.figure(figsize = (10, 10))
plt.hist(obs, bins = 10, )
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of 50 obs with mu=95, sigma=1")
```

```python
xbar = obs.mean()
xbar
```

```python
# setting initial values
mu_init = 95
sigma2_init = 0.3
N = 10000

# conditional distribution 함수 정의
from scipy.stats import invgamma

def mu_given_sigma2(s2):
    # s2는 i-1번째 분산 값
    return norm.rvs(xbar, np.sqrt(s2)/np.sqrt(n))

def sigma2_given_mu(m):
    # m은 i번째 update된 mu값
    beta = sum((obs-m)**2)/2
    return invgamma.rvs(a = n/2, scale = beta, size = 1)
```

```python
### Gibbs Sampling algorithm ###
mu = []
sigma2 = []
mu_old = mu_init
sigma2_old = sigma2_init

for i in range(N):
    mu_new = mu_given_sigma2(sigma2_old)
    sigma2_new = sigma2_given_mu(mu_new)
    mu_old = mu_new
    sigma2_old = sigma2_new
    mu.append(mu_new)
    sigma2.append(sigma2_new)
```

```python
# 앞의 5000개는 burn-in period, 나머지 5000개로 정상분포 추정
mu_burn = mu[5000:] 
sigma2_burn = sigma2[5000:]

plt.figure(figsize = (10, 12))
plt.subplot(211)
plt.hist(np.array(mu_burn), bins = 30, )
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of mu")

plt.subplot(212)
plt.hist(np.array(sigma2_burn), bins = 30, )
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of sigma2")
plt.show()
```

```python
# trace plot
plt.figure(figsize = (10, 10))
plt.subplot(211)
plt.plot(np.linspace(5001, 10000, 5000), np.array(mu_burn))
plt.xlabel("iter")
plt.ylabel("mu")
plt.title("Markov chain by Gibbs sampling")

plt.subplot(212)
plt.plot(np.linspace(5001, 10000, 5000), np.array(sigma2_burn))
plt.xlabel("iter")
plt.ylabel("sigma2")
plt.show()
```
