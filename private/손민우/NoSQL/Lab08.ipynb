```python
from pymongo import MongoClient
```

```python
client = MongoClient()
db = client.lab08
col = db.metro
```

```python
col.find_one()
```

```python
result = col.aggregate([{'$group':{
    '_id':'$line_num',
    'avg_station':{'$avg':'$ride_pasgr_num'},
    'max_station':{'$max':'$ride_pasgr_num'},
    'min_station':{'$min':'$ride_pasgr_num'}
}}])
list(result)
```

```python
result = col.aggregate([{'$match':{'use_dt':{'$gte':'20171221', '$lte':'20171228'}}},
              {'$group':{
                  '_id':{'use_dt':'$use_dt'},
              'total_rider_pasgr':{'$sum':'$ride_pasgr_num'}}},
              {'$sort':{'_id.use_dt':1}}])
list(result)
```
