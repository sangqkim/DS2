```python
from pymongo import MongoClient

client = MongoClient()
db = client.lab4
col = db.grade
```

```python
list(col.find())
```

Print lost document’s sid and Insert the quiz score (80) to the corresponding sid

```python
total_count = (col.count_documents({}) + 1) // 3
# print(total_count)
for sid in range(total_count):
#     count = ...
    #TODO
    count = col.count_documents({'sid':sid})
#     print(sid, count)
    if count < 3:
        missing_id = sid
        print(missing_id, count)
        col.insert_one({'sid':missing_id, 'type':'quiz', 'score':80})
```

print the three students that have the highest quiz score.

```python
list(col.find({'type':'quiz'}).sort([('score', -1)]).limit(3))
```
