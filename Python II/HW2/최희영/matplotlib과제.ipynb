```python
print('Class', df.Class.describe())
print()
print('Class 0:', df[df.Class == 0].Class.count() )
print('Class 1:', df[df.Class == 1].Class.count() )
```

```python
print("Zero")
print(df.Amount[df.Class == 0].describe())
print("\nOne")
print(df.Amount[df.Class == 1].describe())
```

```python
f, (ax1, ax2) = plt.subplots(2,1,sharex=True, figsize=(12,4))
ax1.hist(df.Time[df.Class == 1], bins = 50)
ax1.set_title('Ones')

ax2.hist(df.Time[df.Class == 0], bins = 50)
ax2.set_title('Zero')

plt.xlabel('Time')
plt.ylabel('Count')
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(2,1,sharex=True, figsize=(12,4))
ax1.hist(df.Amount[df.Class == 1], bins = 30)
ax1.set_title('Ones')

ax2.hist(df.Amount[df.Class == 0], bins = 30)
ax2.set_title('Zero')

plt.xlabel('Amount')
plt.ylabel('Number')
plt.yscale('log')
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(2,1,sharex=True, figsize=(12,6))
ax1.scatter(df.Time[df.Class == 1], df.Amount[df.Class == 1])
ax1.set_title('Ones')

ax2.scatter(df.Time[df.Class == 0], df.Amount[df.Class == 0], color='red')
ax2.set_title('Zero')

plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```

```python
plt.scatter(df.Time[df.Class == 0], df.Amount[df.Class == 0], color = 'blue')
plt.scatter(df.Time[df.Class == 1], df.Amount[df.Class == 1], color = 'red')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```
