### 병렬그래프

```python
N=5
menMeans=(20,35,30,35,27)
womenMeans=(25,32,34,20,27)

ind=np.arange(N) #array([0,1,2,3,4])
width=0.35

p1=plt.bar(ind-width/2, menMeans, width, color='#d62728')
p2=plt.bar(ind+width/2, womenMeans, width)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1','G2','G3','G4','G5')) #축 좌표는 -ticks #원래 좌표와 새로운 좌표명 ','로 연결해
plt.yticks(np.arange(0,51,10))
plt.legend((p1,p2),('Men','Women')) #원래 범례와 새로운 범례명도 ','로 연결해
```

### Stack그래프

```python
N=5
menMeans=(20,35,30,35,27)
womenMeans=(25,32,34,20,27)

ind=np.arange(N) #array([0,1,2,3,4])
width=0.35

p1=plt.bar(ind, menMeans, width, color='#d62728')
p2=plt.bar(ind, womenMeans, width, bottom=menMeans) ##요기만 달라집니다. bottom='bar명'

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1','G2','G3','G4','G5')) #축 좌표는 -ticks #원래 좌표와 새로운 좌표명 ','로 연결해
plt.yticks(np.arange(0,81,10))
plt.legend((p1,p2),('Men','Women')) 
```

# 4. pie()
- pie 차트에 들어가는 data의 합은 100%임 (사용자가 제공해야 함)

```python
labels = 'frog' ,'hogs','dogs','logs'
sizes=[15,30,45,10]
explode=(0, 0.2, 0, 0) #explode라는 variables #hogs만 값이 주어져 있어서 걔만 튀어나옴

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
plt.axis('equal') #x축과 y축의 비율이 같다 -> 완전 동그라미
plt.legend()
```

```python
labels = ['frog' ,'pig', 'dog', 'log']
sizes=[15,30,45,10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode=(0, 0.2, 0, 0) #explode라는 variables #hogs만 값이 주어져 있어서 걔만 튀어나옴

plt.pie(sizes, explode=explode, labels=labels,colors=colors, autopct='%1.1f%%',shadow=True, startangle=90)
plt.axis('equal') #x축과 y축의 비율이 같다 -> 완전 동그라미
plt.legend()
```

```python
size=0.3
vals = np.array([[60., 32.],[37.,40.],[29.,10.]])

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.array([1,2,5,6,9,10]))

plt.pie(vals.sum(axis=1), radius=1, colors=outer_colors, wedgeprops=dict(width=size, edgecolor='w'))
plt.pie(vals.flatten(), radius=1-size, colors=inner_colors, wedgeprops=dict(width=size, edgecolor='w'))

plt.axis('equal')
```

# 6. boxplot()
- 통계의 기본함수: mean, 1Q(25%), 3Q(75%), outlier등이 표현됨

```python
spread=np.random.rand(50)*100
center=np.ones(25)*50
flier_high=np.random.rand(10)*100+100
flier_low=np.random.rand(10)*-100
data=np.concatenate((spread, center, flier_high, flier_low),0)
plt.boxplot(data,1) #'1'이 입력되면서 notched box가 되어서 쉽게 파악가능 (median값을) 
```

```python
plt.boxplot(data,1, 'gD') #'gD'가 입력되면서 outlier들의 색,모양 속성이 지정됨
```

```python
plt.boxplot(data,1, '') #''가 입력되면서 outlier들이 사라짐
```

```python
plt.boxplot(data,0, 'rs', 0) #마지막에 입력된 '0' 때문에 horizon하게 그려짐
```

# figure() & subplot()
- 한 화면에서 여러개의 그래프를 그릴때 사용
- subplot(row의 수, columns의 수, 지정하는 위치-단순히 순차적)
- ex. 211 / 221 / 122

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1= np.arange(0.0, 5.0, 0.1) #50개
t2= np.arange(0.0, 5.0, 0.02) #250개

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```

```python
plt.figure(1)
plt.subplot(221)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```

```python
plt.figure(1)
plt.subplot(121)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```
