```python
print("Class", df.Class.describe())
print()
print("Class 0:", df[df.Class==0].Class.count())
print("Class 1:", df[df.Class==1].Class.count())
```

```python
print("Zero")
print(df.Amount[df.Class==0].describe())
print("\nZero")
print(df.Amount[df.Class==1].describe())
```

```python
df_0=pd.DataFrame(df[df.Class==0])
df_1=pd.DataFrame(df[df.Class==1])
```

```python
import matplotlib.patches as mpatches

plt.figure(1, (12,4))
plt.subplot(211)
plt.hist(df_1["Time"], bins=100)
labels = mpatches.Patch(color='blue', label='One')
plt.legend(handles=[labels])

plt.subplot(212)
plt.hist(df_0["Time"], bins=100, color='r')
labels = mpatches.Patch(color='red', label='Zeros')
plt.legend(handles=[labels])

plt.xlabel("Time")
plt.ylabel("Count")
```

#### 2번 plot (Class=1 과 Class=0의 Time data의 hist)
+  Class 1에 속하는 Data는 Time 전체에 걸쳐 발생 분포가 고른 편 (max Count 값이 <40)
+  Class 0에 속하는 Data는 대략 Time 60000, Time 13000에 분포가 집중되어 있음 

```python
plt.figure(2, (12,4))
plt.subplot(211)
plt.hist(df_1["Amount"], bins=30)
plt.xlim(-1000, 27000)
labels = mpatches.Patch(color='blue', label='One')
plt.legend(handles=[labels])

plt.subplot(212)
plt.hist(df_0["Amount"], bins=30, color='r')
labels = mpatches.Patch(color='red', label='Zeros')
plt.legend(handles=[labels])

plt.xlabel("Amount")
plt.ylabel("Count")
plt.yscale('log')
plt.xlim(-1000, 27000)

print("'Ones'의 Amount값의 평균:%d, 표준편차:%d" %(df_1.Amount.mean(), df_1.Amount.std()))
print("'Zero'의 Amount값의 평균:%d, 표준편차:%d" %(df_0.Amount.mean(), df_0.Amount.std()))
```

#### 3번 plot (Class=1 과 Class=0의 Amount hist)
+ 두 class 모두 Amout가 증가할수록 발생량은 감소하는 그래프를 보임.
+ 단, 'Ones' class의 Amount의 평균값이 'Zero'의 Amount 평균값 보다 높으며 표준편차는 두 class가 거의 비슷함.

```python
plt.figure(3, (12,6))
plt.subplot(211)
plt.scatter(df_1["Time"],df_1["Amount"], marker='2')
labels = mpatches.Patch(color='blue', label='One')
plt.legend(handles=[labels])

plt.subplot(212)
plt.scatter(df_0["Time"],df_0["Amount"], c='red', marker='2')
labels = mpatches.Patch(color='red', label='Zeros')
plt.legend(handles=[labels])

plt.xlabel("Time")
plt.ylabel("Amount")
```

```python
plt.figure(4)

p0=plt.scatter(df_0["Time"],df_0["Amount"], c='red', marker='2')
p1=plt.scatter(df_1["Time"],df_1["Amount"], c='blue', marker='2')
plt.title("Ones & Zeros scatter plot", loc='center')
labels = mpatches.Patch(color='blue', label='One')
plt.legend((p0,p1),("Zero", "One"))


plt.xlabel("Time")
plt.ylabel("Amount")
```

#### 4번 plot (Class=1 과 Class=0의 Scatter plot)
+ 추정하기에는 Zero가 머신러닝의 training data이고 One이 test data일 듯

```python
v_features=df.iloc[:,1:29].columns
plt.figure(5, (12,28*4))
gs=gridspec.GridSpec(28,1)
for i, cn in enumerate(df[v_features]):
    ax= plt.subplot(gs[i])
    p1=sns.distplot(df[cn][df.Class ==1], bins=50, color='blue')
    p0=sns.distplot(df[cn][df.Class ==0], bins=50, color='red')
    ax.set_xlabel('')
    ax.set_title('histogram of feature: ' +str(cn))
plt.show()
```

#### 각 V1 ~ V28의 histogram graph (One vs Zero)
+ V13, V15, V22은  유사한 값을 갖는 그래프 결과값을 보임.
+ Class 0은 training data 이고 class 1은 test data이고 V는 각 머신러닝 기법을 통한 테스트 결과라고 보면 V13, V15, V22가 유의미한 결과
