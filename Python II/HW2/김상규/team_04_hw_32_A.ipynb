```python
f, (ax1, ax2) = plt.subplots(2,1,sharex=True, figsize=(12,4))
ax1.hist(df.Time[df.Class == 1], bins=50)
ax1.set_title('ones')
ax2.hist(df.Time[df.Class == 0], bins=50)
ax2.set_title('zeros')
```

```python
f, (ax1, ax2) = plt.subplots(2,1,sharex=True, figsize=(12,4))
ax1.hist(df.Amount[df.Class == 1], bins=30)
ax1.set_title('ones')
ax2.hist(df.Amount[df.Class == 0], bins=30)
ax2.set_title('zeros')
plt.yscale('log')
```

```python
f, (ax1, ax2) = plt.subplots(2,1,sharex=True, figsize=(12,6))
ax1.scatter(df.Time[df.Class == 1], df.Amount[df.Class == 1], color='blue')
ax2.scatter(df.Time[df.Class == 0], df.Amount[df.Class == 0], color='red')
```

```python
v_features = df.iloc[:, 1:29].columns
```

```python
plt.figure(figsize=(12,28*4))
gs = gridspec.GridSpec(28,1)
for i, cn in enumerate(df[v_features]):
    ax = plt.subplot(gs[i])
    sns.distplot(df[cn][df.Class==1], bins=50, color='red')
    sns.distplot(df[cn][df.Class==0], bins=50, color='green')
    ax.set_xlabel('')
    ax.set_title('histogram of features: ' + str(cn))
```
