#### 시간 컬럼이 있으므로 다른 attribute들의 시간에 따른 변화 관찰
- 시간에 따른 컬럼의 변화에는 큰 특징이 없음

```python
fig, h = plt.subplots(5, 6, figsize=(10,10), sharex=True, sharey=False)
for i in range(5):
    for j in range(6):
        h[i, j].plot(df.iloc[:, 0], df.iloc[:, i*6+(j+1)])
        h[i, j].set_title(df.columns[i*6+(j+1)])
```

#### 컬럼들간의 correlation 분석
- amount와 Class 사이에서만 약간의 correlation 존재

```python
df_corr = df.corr()
ax = sns.heatmap(abs(df_corr), vmin=0, vmax=0.4)
plt.show()
```

#### Class별 갯수 및 비율
- Class가 1인 경우는 0.17% 수준으로 매우 적은 비율을 가지고 있음

```python
df.Class.count() # 전체 클래수 갯수 중
df['Class'].sum() # Class == 1인 갯수
df['Class'].sum() / df.shape[0] * 100 # Class == 1 비율
```

#### 각 컬럼별 값 분포
- histogram으로 각 컬럼별 분포를 확인

```python
fig, h = plt.subplots(5, 6, figsize=(15,10), sharex=False, sharey=False)
for i in range(5):
    for j in range(6):
        h[i, j].hist(df.iloc[:, i*6+(j+1)], bins=50)
        h[i, j].set_title(df.columns[i*6+(j+1)])
```

#### Class 값에 따른 컬럼별 분포
- class가 나타나는 비율이 서로 상이하므로 normalized histogram 분석 필요
- class에 따라 feature (column)의 histogram이 다른 것을 확인할 수 있음
- 즉, feature를 이용하여 class를 구분할 수 있음

```python
v_features=df.iloc[:,1:29].columns
plt.figure(5, (12,28*4))
gs=gridspec.GridSpec(28,1)
for i, cn in enumerate(df[v_features]):
    ax= plt.subplot(gs[i])
    p1=sns.distplot(df[cn][df.Class ==1], bins=50, color='blue')
    p0=sns.distplot(df[cn][df.Class ==0], bins=50, color='red')
    ax.set_xlabel('')
    ax.set_title('histogram of feature: ' +str(cn))
plt.show()
```
