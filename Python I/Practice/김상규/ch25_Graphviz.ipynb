```python
# p14 ~ 16
from graphviz import Digraph
# Digraph: 방향성이 있는 그래프
dot = Digraph(comment = 'The Round Table') # class라서 대문자로 시작함
# 방향그래프 클래스 생성

dot.node('A', 'King Arthur') # 노드 추가 (name, label)
dot.node('B', 'Sir Bedevere the Wise')
dot.node('L', 'Sir Lacelot the Brave')

dot.edges(['AB', 'AL']) # A -> B, A -> L인 edge들 추가
dot.edge('B', 'L', constraint = 'false') # B -> L인 edge 하나 추가
print(dot.source)
dot.render('test_graph.gv', view=True) # True인 경우, 저장된 그림파일을 새창으로 불러옴
```

## Finite State Machine

```python
# p46
from graphviz import Digraph

f = Digraph('finite_state_machine', filename='test_graph_fsm.gv')
f.attr(rankdir='LR', size='8.5')

f.attr('node', shape='doublecircle')
f.node('LR_0')
f.node('LR_3')
f.node('LR_4')
f.node('LR_8')

f.attr('node', shape='circle')
f.edge('LR_0', 'LR_2', label='SS(B)')
f.edge('LR_0', 'LR_1', label='SS(S)')
f.edge('LR_1', 'LR_3', label='SS($end)')
f.edge('LR_2', 'LR_6', label='SS(b)')
f.edge('LR_2', 'LR_5', label='SS(a)')
f.edge('LR_2', 'LR_4', label='SS(A)')
f.edge('LR_5', 'LR_7', label='SS(b)')
f.edge('LR_5', 'LR_5', label='SS(a)')
f.edge('LR_6', 'LR_6', label='SS(b)')
f.edge('LR_6', 'LR_5', label='SS(a)')
f.edge('LR_7', 'LR_8', label='SS(b)')
f.edge('LR_7', 'LR_5', label='SS(a)')
f.edge('LR_8', 'LR_6', label='SS(b)')
f.edge('LR_8', 'LR_5', label='SS(a)')

f.view()
```
